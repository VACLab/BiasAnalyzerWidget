```python
# IMPORT BIASANALYZER COMPONENTS
from biasanalyzer.api import BIAS
from CohortViewer import *
from CohortDefinition import (ConditionOccurrence, VisitOccurrence, DateEvent, Demographics, CohortCriteria, AND, OR, BEFORE, NOT,)

# CREATE INSTANCE OF BIAS & CONNECT TO DATA SOURCE
bias = BIAS()
bias.set_config("../config_duckdb_real.yaml")
try:
    bias.set_root_omop()
except:
    bias.cleanup()
    raise  # continue to raise the error
```

```python
hf = ConditionOccurrence(event_concept_id=316139)
hf_def = CohortCriteria(
    temporal_blocks=[hf],
)
# print(hf_def.to_yaml(sort_keys=False))
```

```python
hf_cohort = bias.create_cohort(
    'Heart Failure',
    'All heart failure',
    hf_def,
    'system'
)
# print(f'hf_cohort id = {hf_cohort.cohort_id}')
```

```python
hf_view = CohortViewer(cohort1=hf_cohort, cohort1_shortname="hf")
hf_view
```

```python
# disconnect from the database
bias.cleanup()
```
